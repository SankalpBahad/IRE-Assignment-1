### 2.3 Term Frequency-Inverse Documents Frequency
<font size="3">

The <b><i>tf-idf</b></i> is a weight used in information retrieval and text mining. This weight is a statistical measure used to evaluate how important a word is for a document in a collection or corpus. The importance increases proportionally to the number of times a word appears in the document but is offset by the frequency of the word
in the corpus. Variations of the <b><i>tf-idf</b></i>  weighting scheme are often used by search engines as a central tool in scoring and ranking a documents relevance given a user query.
The <b><i>term frequency</b></i> $tf_{ij}$ is the weight of a term $t_{i}$ in a document $d_{j}$ computed according to
\begin{equation}
tf_{ij}= \frac{freq_{ij}}{\max\limits_{k} freq_{kj}},
\end{equation}
where $freq_{ij}$ is the number of occurrences of the term $t_{i}$ in the document $d_{j}$. As a result, the most frequent term $t_{\bar{i}}$
in a document $d_{j}$ always has $tf_{\bar{i}j} = 1$.
    
The <b><i>inverse document frequency</b></i> $idf_{i}$ is a measure of the general importance of the term $t_{i}$ in
the whole corpus. It is obtained by dividing the number of all documents by the number of
documents containing the term  $t_{i}$, and then taking the logarithm of that quotient as
\begin{equation}
idf_{i}= \log \frac{N}{n_{i}},
\end{equation}
where $N$ is the total number of documents, and $n_{i}$ is the number of documents in which the
term $t_{i}$ appears.
Using the <b><i>term frequency</b></i> $tf_{ij}$ and the <b><i>inverse document frequency</b></i>  $idf_{i}$ as described above, we obtain the most used term-weighting scheme in text
retrieval defned as
\begin{equation}
w_{ij}= tf_{ij} \cdot idf_{i}.
\end{equation}
A high weight $w_{ij}$ is reached by a high frequency of the term $t_{i}$ in the document $d_{j}$ and a low frequency of the term $t_{i}$ in the whole collection of documents. Hence, the weights will tend to filter out common terms that appear in many documents in the collection.

</font>

### 2.4 BM25
<font size="3">

**BM25**, also known as Okapi BM25, is a ranking function used by search engines to estimate the relevance of documents to a given search query. It improves upon the classic TF-IDF model by incorporating document length normalization and tunable parameters. 

**Formulation:**

The BM25 score for a document *D* and query *Q* is calculated as:

\begin{equation}
Score(D, Q) = \sum_{i=1}^{n} IDF(q_i) \cdot \frac{f(q_i, D) \cdot (k_1 + 1)}{f(q_i, D) + k_1 \cdot (1 - b + b \cdot \frac{|D|}{avgdl})}
\end{equation}

Where:

*   **$q_1$, ..., $q_n$** are the query terms.
*   **$f(q_i, D)$** is the term frequency of $q_i$ in document *D*.
*   **$|D|$** is the length of document *D*.
*   **$avgdl$** is the average document length in the collection.
*   **$k_1$** and **$b$** are tunable parameters:
    *   **$k_1$** controls the impact of term frequency.
    *   **$b$** controls the impact of document length.

**IDF(qi)** is calculated similarly to TF-IDF:

\begin{equation}
IDF(q_i) = log \frac{N - n(q_i) + 0.5}{n(q_i) + 0.5}
\end{equation}

Where:

*   **$N$** is the total number of documents in the corpus.
*   **$n(q_i)$** is the number of documents containing $q_i$.

</font> 

### 2.5 Boolean and Vector model

<font size="3">

The <b><i>boolean model</b></i> is a simple retrieval model based set theory on <b><i>boolean algebra</b></i>. The concept of a set is intuitive, <b><i>index terms</b></i> and queries are specified as <b><i>boolean</b></i> expressions. The <b><i>boolean model</b></i> provides easy tools that are easy to be grasped by a common user of an information retrieval system. 
The <b><i>vector model</b></i> represents documents and queries as vectors. The dimensionality of each vector is fixed, so that we can compare similarity of each document $d_{j}$ and each query $q_{l}$ based on the <b><i>cosine of the angle</b></i> between two vectors that represent them. It is expressed according to following rules
\begin{equation}
sim(d_{j},q_{l})=\frac{\langle\vec{d_{j}},\vec{q_{l}}\rangle}{||\vec{d_{j}}||\cdot||\vec{q_{l}}||}.
\end{equation}

Each dimension corresponds to a unique term chosen in advance. Its value describes the significance of the term with respect the document. In the simplest case, it can be the number of times the term occurs in the document. The <b><i>vector model</b></i> disregards the order in which the terms appear in a document.

```python
import os
import re
from nltk.stem import PorterStemmer
from tqdm import tqdm
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import math
from sklearn.metrics.pairwise import cosine_similarity
```

```python
def extract_content_and_clean(directory_path, stopwords, remove_stopwords=True):
    data = []

    for filename in tqdm(os.listdir(directory_path)):
        filepath = os.path.join(directory_path, filename)
        with open(filepath, "r", encoding='utf-8') as f:
            content = f.read()
            content = re.sub(r'[^a-zA-Z\s]', '', content)
            content = re.sub('\n', '', content)
            content = re.sub('  ', ' ', content)
            words = content.split(' ')
            if remove_stopwords:
                words = [word for word in words if word not in stopwords]
            stemmed_words = [stemmer.stem(word.lower()) for word in words]
            data.append({'filename': filename, 'words': stemmed_words})

    df = pd.DataFrame(data)
    return df

stemmer = PorterStemmer()

with open("english.stop", "r") as f:
    stopwords=[i.strip() for i in f.readlines()]

directory_path = "./documents/"

# cleaned_documents_with_stopwords = extract_content_and_clean(directory_path, stopwords, False)

# cleaned_documents_without_stopwords = extract_content_and_clean(directory_path, stopwords, True)
```

### TF-IDF

```python
data=pd.read_csv("cleaned_data_with_stopwords.csv")

all_docs = {}
sum_doc_len = 0
unique_word_counts = {}
all_unique_words_dict = {}
freq = {}

for idx, entry in zip(data['filename'], data['words']):
    tmp = [i.strip()[1:-1] for i in entry[1:-1].split(',')]
    all_docs[idx] = tmp
    sum_doc_len += len(tmp)

avg_doc_len = sum_doc_len / len(data['filename'])

for f in all_docs:
    doc = all_docs[f]
    tmp={}
    for word in doc:
        if word in tmp:
            tmp[word]+=1
        else: tmp[word]=1
    unique_word_counts[f]=tmp

for f in unique_word_counts:
    words=unique_word_counts[f].keys()
    for word in words:
        all_unique_words_dict[word]=1

all_unique_words=list(all_unique_words_dict.keys())


for word in all_unique_words:
    freq[word] = 0
for d in unique_word_counts:
    for word in unique_word_counts[d]:
        freq[word] += 1

def calculate_tf(data, all_docs, unique_word_counts):
    tf={}
    for f in unique_word_counts:
        tmp={}
        for word in unique_word_counts[f]:
            tmp[word]=unique_word_counts[f][word]/len(all_docs[f])
        tf[f]=tmp
    return tf

def calculate_idf(data, all_docs, unique_word_counts, freq):
    total_docs=len(all_docs)
    idf={}
    for word in freq:
        idf[word]=math.log(total_docs/freq[word])
    return idf
```

```python
doc_matrix={}

doc_tf=calculate_tf(data, all_docs, unique_word_counts)
doc_idf=calculate_idf(data, all_docs, unique_word_counts, freq)
```

```python
for f in doc_tf:
    tmp={}
    for word in doc_tf[f]:
        tmp[word]=doc_tf[f][word]*doc_idf[word]
    doc_matrix[f]=tmp
```

### BM25

```python
with open ("queries.txt", "r") as f:
    queries=[i.strip() for i in f.readlines()]

for i in range(len(queries)):
    queries[i] = re.sub(r'[^a-zA-Z\s]', '', queries[i])
    queries[i] = re.sub('\n', '', queries[i])
    queries[i] = re.sub('  ', ' ', queries[i])
    words = queries[i].split(' ')
    queries[i] = [stemmer.stem(word.lower()) for word in words]
```

```python
def calculate_bm25(doctf, docidf, all_docs, unique_word_counts, freq, avgdoclen, data, queries, k, b):
    newidf={}
    total_docs=len(data['filename'])
    for q in queries:
        for word in q:
            if word in docidf: value=freq[word]
            else: value=0
            newidf[word]=math.log((total_docs-value+0.5)/(value+0.5)+1)
    bm25={}
    for f in unique_word_counts:
        tmp={}
        id=0
        ratio=len(all_docs[f])/avgdoclen
        for q in queries:
            id+=1
            score=0
            for w in q:
                if w in doctf[f]: val=doctf[f][w]
                else: val=0
                score+=((newidf[w]*val*(k+1))/(val+k*(1-b+(b*ratio)))) 
            tmp[id]=score
        bm25[f]=tmp
    return bm25
```

```python
bm25=calculate_bm25(doc_tf, doc_idf, all_docs, unique_word_counts, freq, avg_doc_len, data, queries, 1.25, 0.75)
```

```python
def calculate_query_tf(queries, unique_word_counts, query_unique_word_counts):
    tf_query={}

    for id, q in zip(query_unique_word_counts.keys(), query_unique_word_counts.values()):
        tmp={}
        for w in q:
            tmp[w]=query_unique_word_counts[id][w]/len(queries[id])
        tf_query[id]=tmp
    return tf_query

def get_query_idf(queries, query_freq):
    idf_query={}

    for word in query_freq:
        idf_query[word]=math.log(len(queries)/query_freq[word])
    return idf_query
```

```python
data_without_stopwords=pd.read_csv("cleaned_data_without_stopwords.csv")

all_docs_without_stopwords = {}
sum_doc_len_without_stopwords = 0
unique_word_counts_without_stopwords = {}
all_unique_words_dict_without_stopwords = {}
freq_without_stopwords = {}

for idx, entry in zip(data_without_stopwords['filename'], data_without_stopwords['words']):
    tmp = [i.strip()[1:-1] for i in entry[1:-1].split(',')]
    all_docs_without_stopwords[idx] = tmp
    sum_doc_len_without_stopwords += len(tmp)

avg_doc_len_without_stopwords = sum_doc_len_without_stopwords / len(data_without_stopwords['filename'])

for f in all_docs_without_stopwords:
    doc = all_docs_without_stopwords[f]
    tmp={}
    for word in doc:
        if word in tmp:
            tmp[word]+=1
        else: tmp[word]=1
    unique_word_counts_without_stopwords[f]=tmp

for f in unique_word_counts_without_stopwords:
    words=unique_word_counts_without_stopwords[f].keys()
    for word in words:
        all_unique_words_dict_without_stopwords[word]=1

all_unique_words_without_stopwords=list(all_unique_words_dict_without_stopwords.keys())


for word in all_unique_words_without_stopwords:
    freq_without_stopwords[word] = 0
for d in unique_word_counts_without_stopwords:
    for word in unique_word_counts_without_stopwords[d]:
        freq_without_stopwords[word] += 1

def calculate_tf(data_without_stopwords, all_docs_without_stopwords, unique_word_counts_without_stopwords):
    tf={}
    for f in unique_word_counts_without_stopwords:
        tmp={}
        for word in unique_word_counts_without_stopwords[f]:
            tmp[word]=unique_word_counts_without_stopwords[f][word]/len(all_docs_without_stopwords[f])
        tf[f]=tmp
    return tf

def calculate_idf(data_without_stopwords, all_docs_without_stopwords, unique_word_counts_without_stopwords, freq_without_stopwords):
    total_docs=len(all_docs_without_stopwords)
    idf={}
    for word in freq_without_stopwords:
        idf[word]=math.log(total_docs/freq_without_stopwords[word])
    return idf
```

```python
doc_matrix_without_stopwords={}

doc_tf_without_stopwords=calculate_tf(data_without_stopwords, all_docs_without_stopwords, unique_word_counts_without_stopwords)
doc_idf_without_stopwords=calculate_idf(data_without_stopwords, all_docs_without_stopwords, unique_word_counts_without_stopwords, freq_without_stopwords)
```

```python
for f in doc_tf_without_stopwords:
    tmp={}
    for word in doc_tf_without_stopwords[f]:
        tmp[word]=doc_tf_without_stopwords[f][word]*doc_idf_without_stopwords[word]
    doc_matrix_without_stopwords[f]=tmp
```

### BM25

```python
with open ("queries.txt", "r") as f:
    queries_without_stopwords=[i.strip() for i in f.readlines()]

for i in range(len(queries_without_stopwords)):
    queries_without_stopwords[i] = re.sub(r'[^a-zA-Z\s]', '', queries_without_stopwords[i])
    queries_without_stopwords[i] = re.sub('\n', '', queries_without_stopwords[i])
    queries_without_stopwords[i] = re.sub('  ', ' ', queries_without_stopwords[i])
    words = queries_without_stopwords[i].split(' ')
    words=[word for word in words if word not in stopwords]
    queries_without_stopwords[i] = [stemmer.stem(word.lower()) for word in words]
```

```python
def calculate_bm25(doctf_without_stopwords, docidf_without_stopwords, all_docs_without_stopwords, unique_word_counts_without_stopwords, freq_without_stopwords, avgdoclen_without_stopwords, data_without_stopwords, queries_without_stopwords, k, b):
    newidf={}
    total_docs=len(data_without_stopwords['filename'])
    for q in queries_without_stopwords:
        for word in q:
            if word in docidf_without_stopwords: value=freq_without_stopwords[word]
            else: value=0
            newidf[word]=math.log((total_docs-value+0.5)/(value+0.5)+1)
    bm25={}
    for f in unique_word_counts_without_stopwords:
        tmp={}
        id=0
        ratio=len(all_docs_without_stopwords[f])/avgdoclen_without_stopwords
        for q in queries_without_stopwords:
            id+=1
            score=0
            for w in q:
                if w in doctf_without_stopwords[f]: val=doctf_without_stopwords[f][w]
                else: val=0
                score+=((newidf[w]*val*(k+1))/(val+k*(1-b+(b*ratio)))) 
            tmp[id]=score
        bm25[f]=tmp
    return bm25
```

```python
bm25_without_stopwords=calculate_bm25(doc_tf_without_stopwords, doc_idf_without_stopwords, all_docs_without_stopwords, unique_word_counts_without_stopwords, freq_without_stopwords, avg_doc_len_without_stopwords, data_without_stopwords, queries_without_stopwords, 1.25, 0.75)
```

```python
def boolean_model(p):
    top_p_words_without_stopwords=[i[1] for i in vals_without_stopwords[:p]]
    matrix={}
    for d in unique_word_counts_without_stopwords:
        matrix[d]=[]
        for i in top_p_words_without_stopwords:
            matrix[d].append(0)
        idx=0
        for w in top_p_words_without_stopwords:
            if w in unique_word_counts_without_stopwords[d]:
                matrix[d][idx]=1
            idx+=1
    q_matrix={}
    idx=0
    for q in queries_without_stopwords:
        q_matrix[idx]=[]
        for i in top_p_words_without_stopwords:
            q_matrix[idx].append(0)
        idx2=0
        for w in top_p_words_without_stopwords:
            if w in q:
                q_matrix[idx][idx2]=1
            idx2+=1
        idx+=1
    
    retrievals={}
    for i, v1 in zip(q_matrix.keys(), q_matrix.values()):
        retrievals[i]=[]
        for j, v2 in zip(matrix.keys(), matrix.values()):
            if(v1==v2): retrievals[i].append(j)
    
    print(p, end='\t')
    for i in retrievals:
        print(len(retrievals[i]), end="\t")
    print()
    
```

```python
def calculate_query_tf(queries_without_stopwords, unique_word_counts_without_stopwords, query_unique_word_counts_without_stopwords):
    tf_query={}

    for id, q in zip(query_unique_word_counts_without_stopwords.keys(), query_unique_word_counts_without_stopwords.values()):
        tmp={}
        for w in q:
            tmp[w]=query_unique_word_counts_without_stopwords[id][w]/len(queries_without_stopwords[id])
        tf_query[id]=tmp
    return tf_query

def get_query_idf(queries_without_stopwords, query_freq_without_stopwords):
    idf_query={}

    for word in query_freq_without_stopwords:
        idf_query[word]=math.log(len(queries_without_stopwords)/query_freq_without_stopwords[word])
    return idf_query
```

```python
query_matrix_without_stopwords={}
tf_query_without_stopwords=calculate_query_tf(queries_without_stopwords, unique_word_counts_without_stopwords, query_unique_word_counts_without_stopwords)
idf_query_without_stopwords=get_query_idf(queries_without_stopwords, query_freq_without_stopwords)

for q in query_unique_word_counts_without_stopwords:
    tmp={}
    for word in tf_query_without_stopwords[q]:
        tmp[word]=tf_query_without_stopwords[q][word]*idf_query_without_stopwords[word]
    query_matrix_without_stopwords[q]=tmp
```
